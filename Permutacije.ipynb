```python
skup = (1,2,3,4)
l = list(skup)

def permutacije(l):
    if len(l) == 1:
        yield l
    else: 
        for p in permutacije(l[1:]):
            for i in range (len(l)):
                yield p[:i] + l[0:1] + p[i:]

r = permutacije(l)
z = list(r)
print(z)
```
